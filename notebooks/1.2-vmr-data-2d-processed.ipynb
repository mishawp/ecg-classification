# GAF, RP, MTF преобразования

```python
import os
import numpy as np
import wfdb
from pathlib import Path
from dotenv import load_dotenv
import matplotlib.pyplot as plt

load_dotenv()
# os.sys.path.append(os.sys.path.append(os.getenv("PYTHONPATH")))
from src.data.processing2d import process2d

PATH_RAW_DATA = Path(os.getenv("project_root"), "data", "raw", "ptbxl")
PATH_FILE_NORM = Path(PATH_RAW_DATA, "records100", "00000", "00001_lr")
PATH_FILE_NDT = Path(PATH_RAW_DATA, "records100", "15000", "15123_lr")

signal_norm, _ = wfdb.rdsamp(PATH_FILE_NORM)
signal_norm = signal_norm[:, [0, 1, 7]]
imgs_norm = process2d(signal_norm)

signal_imi, _ = wfdb.rdsamp(PATH_FILE_NDT)
signal_imi = signal_imi[:, [0, 1, 7]]
imgs_imi = process2d(signal_imi)

plt.figure(figsize=(12, 16))

plt.subplot(3, 2, 1)
plt.imshow(imgs_norm[0, :, :])
plt.title("NORM. Lead I", fontsize=16)
plt.ylabel("GAF", fontsize=16)

plt.subplot(3, 2, 3)
plt.imshow(imgs_norm[1, :, :])
plt.ylabel("RP", fontsize=16)

plt.subplot(3, 2, 5)
plt.imshow(imgs_norm[2, :, :])
plt.ylabel("MTF", fontsize=16)

plt.subplot(3, 2, 2)
plt.imshow(imgs_imi[0, :, :])
plt.title("IMI, AMI, PVC. Lead I", fontsize=16)

plt.subplot(3, 2, 4)
plt.imshow(imgs_imi[1, :, :])

plt.subplot(3, 2, 6)
plt.imshow(imgs_imi[2, :, :])

plt.tight_layout()
plt.show()
```
