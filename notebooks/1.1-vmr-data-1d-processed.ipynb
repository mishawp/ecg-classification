```python
import os
import numpy as np
import wfdb
from pathlib import Path
from dotenv import load_dotenv
import matplotlib.pyplot as plt

load_dotenv()
# os.sys.path.append(os.sys.path.append(os.getenv("PYTHONPATH")))
from src.data.processing1d import process1d

PATH_RAW_DATA = Path(os.getenv("project_root"), "data", "raw", "ptbxl")
PATH_FILE_NORM = Path(PATH_RAW_DATA, "records100", "00000", "00001_lr")
PATH_FILE_NDT = Path(PATH_RAW_DATA, "records100", "15000", "15123_lr")

signal_norm, _ = wfdb.rdsamp(PATH_FILE_NORM)
signal_norm = signal_norm[:, 0]
processed_signal_norm = process1d(signal_norm.reshape(-1, 1))

signal_imi, _ = wfdb.rdsamp(PATH_FILE_NDT)
signal_imi = signal_imi[:, 0]
processed_signal_imi = process1d(signal_imi.reshape(-1, 1))

#    
wfdb.plot_items(
    signal=np.column_stack((processed_signal_norm, processed_signal_imi)),
    title="  . I ",
    time_units="seconds",
    figsize=(16, 6),
    ecg_grids="all",
    fs=100,
    sig_units=["mV", "mV"],
    sig_name=["NORM", "IMI"],
)
```
